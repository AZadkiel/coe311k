```python
import random
def monteCarloPi(numSample):
    redCounter = 0
    for i in range(0, numSample):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x*x + y*y < 1):
            redCounter = redCounter + 1
    return (4*(redCounter/numSample))
```

Added monteCarloPi from homework/lecture

```python
import numpy
import math
import matplotlib.pyplot as plt

trueErrorList = []
relativeErrorList = []
sampleList = []

for i in range (0,100):
    calculatedPi = monteCarloPi(1000)
    truePi = numpy.pi
    trueError = abs(truePi - calculatedPi)
    relativeError = abs((truePi - calculatedPi)/truePi)
    
    trueErrorList.append(trueError)
    relativeErrorList.append(relativeError)
    sampleList.append(i)
    
    
    print(calculatedPi, trueError, relativeError)
    
plt.subplot(1, 2, 1)
plt.plot(sampleList,trueErrorList)
plt.title("Absolute Error")


plt.subplot(1, 2, 2)
plt.plot(sampleList,relativeErrorList)
plt.title("Relative Error")
```
