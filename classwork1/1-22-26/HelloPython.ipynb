```python
def montecarlo_pi(number_of_shots):
    redCounter = 0
    for i in range(0, number_of_shots):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x*x + y*y < 1):
            redCounter = redCounter + 1
    return (4*(redCounter/number_of_shots))
    
```

```python
montecarlo_pi(10)
```

```python
montecarlo_pi(1000000)
```
