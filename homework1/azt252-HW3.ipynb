```python
import numpy as np
import math


def sinTaylor(numTerm, x):
    sumTaylor = 0
    for i in range(0, numTerm):
        factorialNum = 2*i + 1
        calcTaylor = ((-1)**i)*(x**(2*i + 1))/(math.factorial(factorialNum))
        sumTaylor += calcTaylor #fancy add operator woahhh
    return sumTaylor


print(sinTaylor(4, 10))
print(np.sin(0.2))
```
