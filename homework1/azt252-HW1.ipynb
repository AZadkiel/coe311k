```python
import random
def monteCarloPi(numSample):
    redCounter = 0
    for i in range(0, numSample):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x*x + y*y < 1):
            redCounter = redCounter + 1
    return (4*(redCounter/numSample))

monteCarloPi(1000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def monteCarloSampleDistr(initialV, finalV, steps):
    x = initialV
    xLine = []
    errorLine = []
    while (x <= finalV):
        calcPi = monteCarloPi(x)
        error = abs(calcPi - np.pi)
        xLine.append(x)
        errorLine.append(error)
        x = x + steps
    return xLine, errorLine
    
monteCarloSampleDistr(10, 1000, 20)
```
